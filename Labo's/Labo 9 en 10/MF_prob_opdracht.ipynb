```python
def sample_mean_calculator(input_dist, sample_size, n_samples):
    sample_means = []
    for i in range(n_samples):
        sample = np.random.choice(input_dist, size=sample_size, replace=False)
        sample_means.append(sample.mean())
    return sample_means
```

```python
P1 = 1/52 * 1/52
print(f"Spades in a row: {P1}\n")

P2 = 52/52 * 1/52
print(f"Same card in a row: {P2}\n")

P3 = 1/2 * 1/2 * 1/2 * 2/2 * 2/2
print(f"3 heads out of 5: {P3}\n")

def combinatoric(n, k):
    nFact = n
    for i in range(1, n):
        nFact *= i
    
    kFact = k
    for i in range(1, k):
        kFact *= i
    
    if kFact == 0: kFact = 1
    
    nkFact = n - k
    for i in range(1, n - k):
        nkFact *= i
        
    if nkFact == 0: nkFact = 1

    return nFact / (kFact * nkFact)

mean = 0
for i in range(0, 6):
    mean += combinatoric(5, i)
    print(f"{i} heads: {combinatoric(5, i)}")

mean /= 6
print(mean)
```
